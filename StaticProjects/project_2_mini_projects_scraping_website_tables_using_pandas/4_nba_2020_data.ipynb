In this mini-project, we will scrape the NBA 2020 players data.

URL to be scraped : https://www.basketball-reference.com/leagues/NBA_2020_totals.html

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
URL = 'https://www.basketball-reference.com/leagues/NBA_2020_totals.html'
req = requests.get(URL)
```

```python
soup = BeautifulSoup(req.content, 'html.parser')
```

```python
tables = soup.find(name='table')
```

```python
print("The total number of tables are :", len(tables))
```

```python
tables = pd.read_html(str(tables))
print("Total number of tables are :", len(tables))
```

Let's create a Pandas Dataframe from the table.

```python
nba_data_df = tables[0].set_index('Rk')
print("NBA Players Data (2020) :")
print("-"*200)
nba_data_df
```
