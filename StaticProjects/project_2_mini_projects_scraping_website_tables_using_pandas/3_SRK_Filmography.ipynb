In this mini-project, we will scrape the data for Shah Rukh Khan films.


Url to be scraped : https://en.wikipedia.org/wiki/Shah_Rukh_Khan_filmography

```python
import pandas as pd
URL = "https://en.wikipedia.org/wiki/Shah_Rukh_Khan_filmography"
srk_films_data = pd.read_html(URL)
```

```python
print("Total number of tables :",len(srk_films_data))
```

```python
filmography = srk_films_data[1]
# Assuming your DataFrame is named 'time_table_df'
filmography = filmography.drop(".mw-parser-output .tooltip-dotted{border-bottom:1px dotted;cursor:help}Ref(s)", axis=1)
# Set 'Year' as the index
filmography = filmography.set_index('Year')
# Now, 'Year' is the index and the unwanted column is dropped
print("Yearwise Shah Rukh Khan's filmography :")
print("-"*200)
filmography
```

List of All awards and nominations win by Shah Rukh Khan.

```python
import pandas as pd
URL = "https://en.wikipedia.org/wiki/List_of_awards_and_nominations_received_by_Shah_Rukh_Khan"
srk_awards_data = pd.read_html(URL)
```

```python
print("Total number of tables :", len(srk_awards_data))
```

Awards and nominations recieved by Shah Rukh Khan

```python
awards = srk_awards_data[0]
awards= awards.drop("Ref(s)", axis=1)
# Set 'Year' as the index
print("Awards and nominations received by Shah Rukh Khan :")
print("-"*200)
awards
```

State honours received by Shah Rukh Khan :

```python
state_honours = srk_awards_data[1]
# Dropping the unwanted column
state_honours = state_honours.drop("Ref(s)", axis=1)
state_honours_sorted = state_honours.sort_values(by='Year')
print("State honours received by Shah Rukh Khan sorted by year:")
print("-" * 200)
state_honours_sorted
```
