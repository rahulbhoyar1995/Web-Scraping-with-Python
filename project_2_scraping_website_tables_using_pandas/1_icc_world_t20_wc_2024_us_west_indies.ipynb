### Times of India URL

```python
URL  = "https://timesofindia.indiatimes.com/sports/cricket/icc-mens-t20-world-cup/t20-world-cup-2024-schedule-full-list-of-matches-with-dates-and-venues/articleshow/106580368.cms"

timetables_icc_page = pd.read_html(URL)
```

```python
print("Total number of tables :", len(timetables_icc_page))
```

```python
# Assuming timetables_icc_page[0] is your DataFrame
time_table_df = timetables_icc_page[0]

# Set the first row as column headers
time_table_df.columns = time_table_df.iloc[0]

# Drop the first row (which is now the column headers) to make it the index
time_table_df = time_table_df.drop(0)

# Set the first column as the index
time_table_df = time_table_df.set_index(time_table_df.columns[0])

# Remove rows where all values are null
time_table_df = time_table_df.dropna(how='all')

# Now, time_table_df has the first row as column headers, the first column as the index,
# and rows with all null values removed
time_table_df
```
